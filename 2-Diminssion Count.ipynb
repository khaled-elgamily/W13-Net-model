```python
import os
import cv2

dimensions_count = {}

# Loop through files in the folder
for filename in os.listdir(train_images):
    file_path = os.path.join(train_images, filename)

    # Check if the file is an image (you can add more image extensions as needed)
    if file_path.endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp')):
        # Read the image to get its dimensions
        image = cv2.imread(file_path)
        if image is not None:
            height, width, _ = image.shape

            # Create a string representation of the dimensions (e.g., "widthxheight")
            dimension_key = f"{width}x{height}"

            # Increment the count for this dimension or initialize it if it doesn't exist
            dimensions_count[dimension_key] = dimensions_count.get(dimension_key, 0) + 1

# Print the counts for each unique dimension
for dimension, count in dimensions_count.items():
    print(f"Dimensions: {dimension} - Number of Images: {count}")
```

```python
# Initialize counters
num_images = 0

# Loop through files in the folder
for filename in os.listdir(train_masks):
    file_path = os.path.join(train_masks, filename)

    # Check if the file is an image (you can add more image extensions as needed)
    if file_path.endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp')):
        # Increment the image counter
        num_images += 1

        # Read the image to get its dimensions
        image = cv2.imread(file_path)
        if image is not None:
            height, width, _ = image.shape
            print(f"Image {num_images}: {filename} - Dimensions: {width}x{height}")

# Print the total number of images found
print(f"Total number of images: {num_images}")
```
