```python
import os
import random
import shutil

# Define your data directory paths
image_data_dir = "/content/drive/MyDrive/Colab Notebooks/datasets/Aug/patched_images"
mask_data_dir = "/content/drive/MyDrive/Colab Notebooks/datasets/Aug/patched_masks"

# Define the output directory paths
output_image_dir = "/content/drive/MyDrive/Colab Notebooks/datasets/Aug/Images"
output_mask_dir = "/content/drive/MyDrive/Colab Notebooks/datasets/Aug/Masks"

# Define the split ratios (e.g., 70% train, 15% validation, 15% test)
train_ratio = 0.7
validation_ratio = 0.15

# List all image files in the source directory
image_files = os.listdir(image_data_dir)
random.shuffle(image_files)  # Shuffle the files for randomness
total_files = len(image_files)

# Calculate the split points
train_split = int(total_files * train_ratio)
validation_split = int(total_files * (train_ratio + validation_ratio))

# Divide and copy image and mask files to the respective directories
for i, image_file in enumerate(image_files):
    mask_file = image_file.replace(".jpg", ".png")  # Assuming the mask files have the same names with a different extension

    if i < train_split:
        dest_image_dir = os.path.join(output_image_dir, "Train")
        dest_mask_dir = os.path.join(output_mask_dir, "Train")
    elif i < validation_split:
        dest_image_dir = os.path.join(output_image_dir, "Validation")
        dest_mask_dir = os.path.join(output_mask_dir, "Validation")
    else:
        dest_image_dir = os.path.join(output_image_dir, "Test")
        dest_mask_dir = os.path.join(output_mask_dir, "Test")

    # Create destination directories if they don't exist
    os.makedirs(dest_image_dir, exist_ok=True)
    os.makedirs(dest_mask_dir, exist_ok=True)

    # Copy the image and mask files
    shutil.copy(os.path.join(image_data_dir, image_file), os.path.join(dest_image_dir, image_file))
    shutil.copy(os.path.join(mask_data_dir, mask_file), os.path.join(dest_mask_dir, mask_file))
```
